#### Functions em Iterables

- Exemplo: digamos que eu tenha uma function que corrige um código de um produto (semelhante ao que fizemos na seção de function)

```python
def padronizar_texto(texto):
    texto = texto.casefold()
    texto = texto.replace("  ", " ")
    texto = texto.strip()
    return texto
```

- Agora queremos padronizar uma lista de códigos:

```python
produtos = [' ABC12 ', 'abc34', 'AbC37', 'beb12', ' BSA151', 'BEB23']
```

- Usando o for, temos que percorrer a lista toda e para cada item executar a function

```python
for i, produto in enumerate(produtos):
    produtos[i] = padronizar_texto(produto)
print(produtos)
```

- Usando o map, apenas chamamos a função e ela já faz isso para a gente

```python
produtos = list(map(padronizar_texto, produtos)) #Chama a função e passa do lado a lista que você quer aplicar
print(produtos)
```

### Sort (ou sorted)

```python
produtos = ['apple tv', 'mac', 'Iphone x', 'Iphone 11', 'Ipad', 'apple watch', 'mac book', 'airpods']
produtos.sort()
print(produtos)
```

- Como fariamos para ordenar corretamente? (Sem levar em conta as maiúsculas e minúsculas)

```python
produtos.sort(key=str.casefold)
print(produtos)
```

### Outro exemplo: como ordenar um dicionário de acordo com o valor

```python
vendas_produtos = {'vinho': 100, 'cafeteira': 150, 'microondas': 300, 'iphone': 5500}
```

- Queremos ordenar do maior para o menor, para enviar como report para o diretor, por exemplo

```python
#pra isso vamos transformar esse dicionário em lista, e a melhor forma de fazer isso é criando uma lista de tuplas
def segundo_item(tupla):
    return tupla[1]

lista_vendas = list(vendas_produtos.items()) #items é uma função de dicionário que transforma cada item numa tupla
lista_vendas.sort(key=segundo_item, reverse=True)
print(lista_vendas)
print('-' * 50)  #Ai você pode deixar como lista de tuplas ou se preferir transformar em dicioário novamente, fica a sua disposição
print(dict(lista_vendas))
```

### Lambda Expressions

```python
#Fazendo através de função
def minha_funcao(num):
    return num * 2
print(minha_funcao(5))

#Fazendo através de Lambda Expression
minha_funcao2 = lambda num: num * 2
print(minha_funcao2(5))
```

- Exemplo útil: Vamos usar Lambda Expressions para criar uma função que calcula o preço dos produtos acrescido do imposto

```python
imposto = 0.3
#Fazendo com function
def preco_imposto(preco):
    return preco * (1 + imposto)

preco_imposto2 = lambda preco : preco * (1 + imposto)

print(preco_imposto(100))
print(preco_imposto2(100))
```

### Usar lambda como argumento de alguma outra função, como map e filter

```python
preco_tecnologia = {'notebook asus': 2450, 'iphone': 4500, 'samsung galaxy': 3000, 'tv samsung': 1000, 'ps5': 3000, 'tablet': 1000, 'notebook dell': 3000, 'ipad': 3000, 'tv philco': 800, 'notebook hp': 1700}
```

### map()
- Queremos saber o preço de cada produto adicionando o valor do imposto de 30% sobre o valor do produto

```python
#Fazendo por function
def calcular_preco(preco):
    return preco * 1.3

preco_com_imposto = list(map(calcular_preco, preco_tecnologia.values()))
print(preco_com_imposto)

#Fazendo por Lambda
preco_com_imposto2 = list(map(lambda preco: preco * 1.3, preco_tecnologia.values()))
print(preco_imposto2)
```

### filter()
- Queremos apenas os produtos que custam acima de 2000

filter(funçaõ, iterable) -> retorna como resposta todos os itens do iterable onde a função é True

```python
#fazendo por function
def ehmaior2000(item):
    return item[1] > 2000

produtos_acima200 = dict(filter(ehmaior2000, preco_tecnologia.items()))
print(produtos_acima200)

#A diferença do filter pro map, é que o map faz uma conta, o filter retorna apenas verdadeiro ou falso

#fazendo por Lambda
produtos_acima200_2 = dict(filter(lambda item: item[1] > 2000, preco_tecnologia.items()))
print(produtos_acima200_2)
```

### Lambda Expressions para gerar funções
- Usado dentro da definição de uma outra função

### Exemplo:
Vamos criar uma função que permita calcular o valor acrescido de imposto de diferentes categorias (produto, serviço, royalties, etc.)

```python
def calcular_imposto(imposto):
    return lambda preco : preco * (1 + imposto)
```

- Agora vamos definir as funções que calculam o imposto das 3 categorias (produto, serviço, royalties)

```python
calcular_preco_produto = calcular_imposto(0.1)
calcular_preco_servico = calcular_imposto(0.15)
calcular_preco_royalties = calcular_imposto(0.25)
```

- Agora vamos aplicar com um valor de nota fiscal de 100 para ver o resultado

```python
print(calcular_preco_produto(100))
print(calcular_preco_servico(100))
print(calcular_preco_royalties(100))
```
