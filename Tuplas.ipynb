#### Lista de Tuplas
Exemplo: Um dia extraimos as vendas de um banco de dados e elas vieram nesse formato:

```python
vendas = [
    ('20/08/2020', 'iphone x', 'azul', '128gb', 350, 4000),
    ('20/08/2020', 'iphone x', 'prata', '128gb', 1500, 4000),
    ('20/08/2020', 'ipad', 'prata', '256gb', 127, 6000),
    ('20/08/2020', 'ipad', 'prata', '128gb', 981, 5000),
    ('21/08/2020', 'iphone x', 'azul', '128gb', 397, 4000),
    ('21/08/2020', 'iphone x', 'prata', '128gb', 1017, 4000),
    ('21/08/2020', 'ipad', 'prata', '256gb', 50, 6000),
    ('21/08/2020', 'ipad', 'prata', '128gb', 4000, 5000),
]
```

- Qual foi o faturamento do iPhone no dia 20/08/2020?
- Qual foi o produto mais vendido (em unidades) no dia 21/08/2020?

```python
faturamento = 0

for venda in vendas:
    data, produto, cor, capacidade, unidades, valor_unitario = venda
    if produto == 'iphone x' and data == '20/08/2020':
        faturamento += unidades * valor_unitario

print(f'O faturamento de Iphone no dia 20/08/2020 foi de R${faturamento:,}')
```

```python
unidades_vendidas = 0
produto_maisvendido = '' 

for venda in vendas:
    data, produto, cor, capacidade, unidades, valor_unitario = venda
    if data == '21/08/2020' and unidades > unidades_vendidas:
            produto_maisvendido = produto
            unidades_vendidas = unidades

print('Meu produto mais vendido no dia 21/08/2020 foi o {} com {} unidades'.format(produto, unidades))
```

#### Análise de Vendas

```python
meta = 10000
vendas = [
    ('João', 15000),
    ('Julia', 27000),
    ('Marcus', 9900),
    ('Maria', 3750),
    ('Ana', 10300),
    ('Alon', 7870)
]

for vendedor, qtde in vendas:
    if qtde >= meta:
        print('{} bateu a meta, vendeu {} unidades'.format(vendedor, qtde))
```

#### Comparação com o Ano Anterior

```python
vendas_produtos = [('Iphone', 558147, 951642), ('galaxy', 712350, 244295), ('Ipad', 573823, 26964)]

for produto, vendas2019, vendas2020 in vendas_produtos:
    if vendas2020 > vendas2019:
        print('Produto {} teve {} vendas em 2019, e {} vendas em 2020. Crescimento de {:.0%}'
            .format(produto, vendas2019,vendas2020, vendas2020 / vendas2019 - 1))
```
