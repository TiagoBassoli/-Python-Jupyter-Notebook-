```python
faturamento = 1000
custo = 400
lucro = faturamento - custo

# com format
print('O faturamento da loja foi de {}. E o lucro de {}.'.format(faturamento, lucro))
# com f-string
print(f"O faturamento da loja foi de {faturamento}. E o lucro de {lucro}.")
```

12. Faça um programa que pergunte quanto você ganha por hora e n° de horas trabalhadas por mês

```python
GanhoHora = float(input('Quanto você ganha pr hora?'))
HorasTrabalhadas = float(input("Quantas horas você trabalha?"))

Salario_Bruto = GanhoHora * HorasTrabalhadas
print(f"Seu salário bruto é de {Salario_Bruto}")

#desconto de 11% de IR
Salario_depoisdeRoubado = Salario_Bruto * 0.89
print(f'Seu salário depois de roubado pela Receita Federal é {Salario_depoisdeRoubado}')

#desconto do INSS 
desconto_INSS = Salario_Bruto * 0.92
print(f'Seu salário depois de ser alocado em um esquema de pirâmide é {desconto_INSS}')

#desconto do sindicato
desconto_sindicato = Salario_Bruto * 0.95
print(f"Salário após o desconto do sindicato é {desconto_sindicato}")

#Salario Liquido
salario_liquido = Salario_Bruto * 0.76
print(f'Seu salário depois de todos os descontos é {salario_liquido}, boa sorte guerreiro!')
```

13. Programa para uma loja de tintas. Tamanho em metros quadrados da área a ser pintada. Cobertura da tinta é de 1L para cada 3m2, e vendida em latas de 18L, custam R$80,00.

```python
TamanhoaserPintado = 54
LitroporArea = 3
Lata = 18
Preco = 80

Quantidade_comprada = (Lata * LitroporArea) / TamanhoaserPintado
Preco_final = Quantidade_comprada * Preco
print(f'Você deve comprar {Quantidade_comprada} lata(s) para todo o serviço!')
print(f'E o preço vai sair por {Preco_final} reais')
```

14. Programa que pede o tamanho de um arquivo em MB e a velocidade de um link de internet em Mbps, calcule e informe o tempo aproximado de download do arquivo nesse link

```python
Tamanho_Arquivo = float(input('Qual o tamanho do arquivo?'))
Velocidade = float(input('Velocidade de download'))

Tempo = (Tamanho_Arquivo / Velocidade) /60
print(f'Vai demorar {Tempo} minutos para baixar o arquivo')
```
