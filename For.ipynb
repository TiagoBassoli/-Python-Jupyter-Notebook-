Enumerate exemplo: Em uma fábrica, você não pode deixar os produtos em falta, tem de ter um nível mínimo de estoque. Ache os produtos que estão abaixo do mínimo

```python
estoque = [1200, 300, 800, 1500, 1900, 2750, 400, 20, 23, 70, 90, 80, 1100, 999, 900, 880, 870, 50, 1111, 120, 300, 450, 800]
produtos = ['coca','pepsi','guarana','skol','brahma','agua','del valle', 'dolly', 'red bull', 'cahaça', 'vinho','vodka','vinho branco', 'tequila', 'champagne']
nivel_minimo = 50

for i,qtde in enumerate(estoque):
    if qtde < nivel_minimo:
        print('{} está abaixo do nível mínimo. Temos apenas {} unidades.'.format(produtos[i],qtde))
```

#### % de Vendedores que Bateu a Meta

```python
meta = 10000
vendas = [
    ['João', 15000],
    ['Julia', 27000],
    ['Marcus', 9900],
    ['Maria', 3750],
    ['Ana', 10300],
    ['Alon', 7870]
]

bateram_meta = 0
num_funcionarios = len(vendas)

print('Temos nessa empresa {} funcionários'.format(num_funcionarios))

for venda in vendas:
    if venda[1] >= meta:
        bateram_meta += 1 
        porcentagem_bateumeta = bateram_meta/num_funcionarios
print(f'Os quais {bateram_meta} bateram a meta de {meta}')
print('O que representa que {:.1%} dos funcionários alcançaram a meta'.format(porcentagem_bateumeta))
```

Quem foi o vendedor que mais vendeu?

```python
melhor_vendedor = ''
maior_valordevendas = 0
for venda in vendas:
    if venda[1] > maior_valordevendas:
        maior_valordevendas = venda[1]
        melhor_vendedor = venda[0]
    else:
        pass
print(f'O maior valor de vendas foi {maior_valordevendas}')
print('Isso foi feito pelo/a vendedor(a) {}'.format(melhor_vendedor))
```

#### Formas de interromper um For

- break -> interrompe e finaliza o for
- continue -> interrompe e vai para o próximo item do For

```python
vendas = [100, 150, 1500, 2000, 120]
meta = 110

#Se todas as vendas forem acima da meta a loja ganha bônus, se uma não for não tem porque verificar todos e pesar o código
for venda in vendas:
    if venda < meta:
        print('A loja não ganha bônus')
        break

#Quem bateu a meta, conitnue vai pulando que não cumpriu o requisito (bater a meta)
vendedores = ['João', 'Julia', 'Ana', 'José', 'Maria']
meta = 130

for i, venda in enumerate(vendas):
    if venda < meta:
        continue
    print(f'O vendedor {vendedores[i]} bateu a meta, vendendo {venda} unidades')
```
