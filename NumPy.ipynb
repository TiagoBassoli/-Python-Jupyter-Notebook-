### Operações Matemáticas

Se você tentar adicionar um número a todos os elementos de uma lista, você receberá um erro.

```python
lista = [1,2,3,4,5]
nova_lista = lista + 1
print(nova_lista)
```

Se eu quisesse fazer isso, teria que ser obrigado a criar um "For" e passar entre cada elemento da lista

```python
nova_lista = []

for numero in lista:
    nova_lista.append(numero + 1)
print(nova_lista)
```

Com um array NumPy, você pode adicionar (ou subtrair, multiplicar, dividir) um número a todos os elementos de uma vez. Bem mais eficiente. Evita criar for

```python
import numpy as np

array = np.array([1,2,3,4,5])
novo_array = array + 1
print(novo_array)
array_mult = array * 10
print(array_mult)
array_quadrado = array **2
print(array_quadrado)
```

### Desempenho

```python
import numpy as np
import time

# crie uma lista e array com 10 milhões de números
lista = list(range(1, 10_000_001))
array = np.array(range(1, 10_000_001))

#calcule a soma de todos os números na Lista
inicio = time.time()
soma_lista = sum(lista)
fim = time.time()
print(f'Tempo para somar todos os números na lista: {fim - inicio} segundos')

#calcule a soma de todos os números no Array
inicio = time.time()
soma_array = np.sum(array)
fim = time.time()
print(f'Tempo para somar todos os números no array: {fim - inicio} segundos')
```

Usada para calcular o produto escalar de dois arrays. Por exemplo, em uma empresa de varejo, você pode querer calcular o valor total de vendas, dado o número de cada produto vendido e o preço de cada produto.

### Cenário onde esses dados aleatórios podem ser úteis:
Vamos supor que a gente seja um analista de vendas em uma empresa e queira entender melhor o desempenho de um produto específico. No entanto, não temos acesso aos dados reais de vendas, então vamos gerar valores aleatórios para análise. Temos que tivermos eles, é só substituir na análise já estruturada

```python
# Gerar dados de vendas falsos para 30 dias 
# Vamos supor que as vendas de um produto podem variar de 50 a 200 por dia
# Pode definir uma seed para garantir que os resultados sejam reproduzíveis.
rng = np.random.default_rng(seed=0) #Toda pessoa que rodar esse código com esse mesmo valor, vai ter os mesmos números
dados_vendas = rng.integers(low=50, high=200, size=30) #Mínimo 50, máximo 200, tamanho 30
print(dados_vendas)
```

Agora, podemos usar esses dados para várias análises. Por exemplo, dia com as vendas mais altas, mais baixas, ou a média de vendas durante o mês

```python
print(f'Máximo de vendas nesse mês: {np.max(dados_vendas)}')
print(f'Dia com mais vendas: {np.argmax(dados_vendas) + 1}') #+1 por causa que estamos contanto dias, e a programação sempre conta a partir do 0, não queremos isso.
print(f'Mínimo de vendas nesse mês: {np.min(dados_vendas)}')
print(f'Dia com menos vendas: {np.argmin(dados_vendas) + 1}')
print(f'Média de vendas nesse mês: {np.mean(dados_vendas)}')
```

### Filtros e np.where()

```python
#Salário dos funcionários
salarios = np.array([3000, 3500, 4000, 2000, 4500, 4000, 5000])

#Calcular a média salarial
media_salarial = np.mean(salarios)

print(media_salarial)

#Identificar funcionários com salários acima da média
funcionarios_acimamedia = np.where(salarios > media_salarial)
print(salarios[funcionarios_acimamedia]) #NumPy também aceita que você passe Arrays como Indice
#Outros métodos dentro do Array
print(salarios[salarios > media_salarial])
print(salarios[salarios >= 4000])

print(np.where(salarios > media_salarial, 'acima da média', 'abaixo da média')) #O Where pode ser bem mais trabalhado, (expressão, condição se verdadeiro, se falso)

salarios_bonus = np.where(salarios < media_salarial, salarios * 1.1, salarios) #Aumento de 10% pra quem tem o salário abaixo da média
print(salarios_bonus)
```

```python
salarios_ajustados = np.where((salarios >= 3000) & (salarios <= 4500), salarios * 1.05, salarios)
print(salarios_ajustados)
```

```python
print(np.where((salarios_ajustados < 3000) | (salarios_ajustados > 4500)))
```

```python
print(np.where((salarios_ajustados < 3000) | (salarios_ajustados > 4500), salarios_ajustados * 1.1, salarios_ajustados))
```

#### Você é um engenheiro de produção e tem os tempos de ciclo (em minutos) de uma linha de produção em um array NumPy. Seu trabalho é identificar quaisquer tempos de ciclo que estão dois desvios padrão acima ou abaixo da média

```python
import numpy as np

tempos_ciclo = np.array([5.5, 5.7, 5.9, 6.0, 5.8, 5.6, 5.7, 7.2, 4.8])
media = np.mean(tempos_ciclo)
desvio_padrao = np.std(tempos_ciclo)
print(media)
print(desvio_padrao)

condicao = (tempos_ciclo > media + 2 * desvio_padrao) | (tempos_ciclo < media - 2 * desvio_padrao)
anomalias = np.where(condicao)
print(tempos_ciclo[anomalias])
# Print mais elaborado
print(f'Os tempos que estão na condição pedida são {tempos_ciclo[anomalias]}')
```

Para pegar o contrário dessa condição (Ou seja, a parte que está entre os 2 desvios padrões acima e abaixo da média) usamos o ~ (negação da condição)

```python
print(~condicao) #Ele simplesmente inverteu 
print(tempos_ciclo[~condicao])

#No Python padrão existe o NOT ao invés de ~ no NumPy (Muitas coisas do Python padrão tem no NumPy, só é mudado a simbologia)
```

Bonus (Usando o normal, Distribuição Normal ou Gaussiana) para curva de probabilidade em forma de sino, o valor central é a média, na esquerda é a média - desvio padrão, na direita, média + desvio padrão

```python
rng = np.random.default_rng(seed=0)
tempos_gerados = rng.normal(loc=media, scale=desvio_padrao, size=10)
print(tempos_gerados)
```

```python
condicao = ((tempos_gerados > tempos_gerados.mean() + 2 * tempos_gerados.std()) | (tempos_gerados < tempos_gerados.mean() - 2 * tempos_gerados.std()))
print(np.where(condicao))
print(tempos_gerados[condicao]) #Não tem dados de desvio padrão 2 pontos acima ou abaixo da média
```

#### np.reshape()
Muda o formato de um array. Por exemplo, se temos dados de vendas para 2 semanas e queremos reorganizar em uma matriz de 2x7 (2 semanas, 7 dias por semana)

```python
import numpy as np
# Vendas diárias para 2 semanas
vendas = np.array([200,220,250,210,300,280,230,210,220,240,230,210,280,220])
# Reorganizar os dados em uma matriz de 2x7 (útil se quisermos analisar cada semana separadamente)
vendas_reshaped = np.reshape(vendas, (2, 7)) #2 dimensões de 7 elementos (2 linhas e 7 colunas)
print(vendas_reshaped)
```

```python
#Array antes do reshape
print(vendas)
print(vendas.ndim) #Uma única dimensão de valores
print(vendas.shape)

print('-' * 50)
#Array depois do reshape
print(vendas_reshaped)
print(vendas_reshaped.ndim)
print(vendas_reshaped.shape)
```

#### Outras operações

```python
print(vendas_reshaped[0])
print(vendas_reshaped[1])
print(vendas_reshaped[0][0])
print(vendas_reshaped[1][0])
print('-'*50)
print(vendas_reshaped.sum())
print(f'Soma do primeiro eixo: {vendas_reshaped.sum(axis=0)}') #Soma empilhando (cada coluna)
print(f'Soma do segundo eixo: {vendas_reshaped.sum(axis=1)}') #Soma da primeira e segunda linha
```

- A média de vendas por dia da semana

### Exercícios NumPy

```python
import numpy as np

vendas = np.array([[50, 60, 70, 65, 80],
                   [85, 90, 78, 92, 88],
                   [72, 75, 68, 77, 76]])

# Calcular as vendas totais para cada produto
total_vendas_produtos = vendas.sum(axis= 1)
print(f'Vendas totais do Produto A: {total_vendas_produtos[0]}')
print(f'Vendas totais do Produto B: {total_vendas_produtos[1]}')
print(f'Vendas totais do Produto C: {total_vendas_produtos[2]}')
# Vendas totais em cada dia
print('-' * 35)
total_vendas_dia = vendas.sum(axis=0)
for i in range(5):
        print(f'Vendas totais para o Dia {i + 1}: {total_vendas_dia[i]}')
```
