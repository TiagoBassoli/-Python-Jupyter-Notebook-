Formatar com format (Moeda, duas casas decimais, etc)

Exemplo de alinhamento

```python
email = 'urmdavdol@gmail.com'
print('Meu e-mail não é {:^30}, show?'.format(email))
#:< alinha o texto na esquerda
#:> alinha o texto na direita
#:^ alinha o texto no centro
```

Exemplo de Edição de sinal

```python
custo = 500
faturamento = 270
lucro = faturamento - custo
print('Faturamento foi {:+} e lucro foi {:+}'.format(faturamento,lucro))
#:+ sempre mostra o sinal do número
```

Exemplo de separador de milhar

```python
custo = 5000
faturamento = 2700
lucro = faturamento - custo
print('Faturamento foi de {:+,}, e o lucro foi de {:+,}'.format(faturamento,lucro))
#:, -> separador de milhar, dá pra juntar com o sinal também
```

Formato com casas decimais fixas

```python
custo = 500 
faturamento = 270
lucro = faturamento - custo
print('Faturamento foi {:.2f} e lucro foi {:.3f}'.format(faturamento,lucro)) 
#:.'quantidade de casas decimais' e 'f' no final (:.3f, :.2f)
```

Formato percentual

```python
custo = 500 
faturamento = 1300
lucro = faturamento - custo
margem = lucro / faturamento
print('Margem de lucro foi de {:.2%}'.format(margem)) 
#.1% -> mostra o número em porcentagem e com quantas casas decimais você definir, nesse caso 1
```

Formato Moeda -> Combinação de formatos

```python
custo = 5000
faturamento = 27000
lucro = faturamento - custo
print('Faturamento foi R${:,.2f} e lucro foi de {:,.2f}'.format(faturamento,lucro))

#transformando no formato brasileiro
lucro_texto = 'R${:_.2f}'.format(lucro)
print(lucro_texto.replace('.', ',').replace('_','.'))
```

Função round -> arredonda números

```python
imposto = 0.15758
preco = 100
valor_imposto = round(preco * imposto, 1)
print('Imposto sobre o preço é {}'.format(valor_imposto))
```

Step

```python
codigo = "1.2.3.4,5,1,2.3.4,7.9"

pedaco_cod = codigo[::2] #Step pula de tantos em tantos (2 em 2, 3 em 3, e assim vai indo)
print(pedaco_cod)

parte_cod = codigo[::-2] #Pulando de trás pra frente
print(parte_cod)

part_cod = codigo [2:10:2] #Começa no 2 e vai até o 9 de 2 em 2
print(part_cod)
```

#### 1. Tamanho de strings. Faça um programa que leia 2 strings e informe o conteúdo delas seguido do seu comprimento. Informe também se as duas strings possuem o mesmo comprimento e são iguais ou diferentes no conteúdo.
<pre>
Compara duas strings
String 1: Brasil Hexa 2006
String 2: Brasil! Hexa 2006!
Tamanho de "Brasil Hexa 2006": 16 caracteres
Tamanho de "Brasil! Hexa 2006!": 18 caracteres
As duas strings são de tamanhos diferentes.
As duas strings possuem conteúdo diferente.
</pre>

```python
string1 = 'Brasil Hexa 2006'
string2 = 'Brasil! Hexa 2006!'

tamanho_string1 = len(string1)
tamanho_string2 = len(string2)

print(f'Tamanho de {string1}: {tamanho_string1} caracteres')
print(f'Tamanho de {string2}: {tamanho_string2} caracteres')

if string1 == string2:
    print("As duas strings são iguais")
else:
    print('As duas strings possuem conteúdo diferente')

if tamanho_string1 == tamanho_string2:
    print('As duas strings são de tamanhos iguais')
else:
    print('As duas strings são de tamanhos diferentes')
```

#### 2. Valida e corrige número de telefone. Faça um programa que leia um número de telefone, e corrija o número no caso deste conter somente 7 dígitos, acrescentando o '3' na frente. O usuário pode informar o número com ou sem o traço separador.
<pre>
Valida e corrige número de telefone
Telefone: 461-0133
Telefone possui 7 dígitos. Vou acrescentar o digito três na frente.
Telefone corrigido sem formatação: 34610133
Telefone corrigido com formatação: 3461-0133
</pre>

```python
telefone = '461-0133'

telefone = telefone.replace('-', '')
if len(telefone) == 7:
    print('Telefone possui 7 dígitos. Vou acrescentar o digito três na frente')
    telefone = '3' + telefone
    print(f'Telefone corrigido sem formatação: {telefone}')
    print(f'Telefone corrigido com formatação: {telefone[:4]}-{telefone[:4]}')
else: 
    print('O telefone não tem 7 dígitos')
```
