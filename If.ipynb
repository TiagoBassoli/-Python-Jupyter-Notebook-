Mini sistema de controle de estoque

```python
produto = input('Qual o produto?')
categoria = input('Qual a categoria?')
qtde = input('Qual a quantidade atual de produtos?')

if produto and categoria and qtde:
    qtde = int(qtde)
    if categoria == 'bebidas':
        if qtde < 75:
            print(f"Solicitar {produto} à equipe de compras, tempos apenas {qtde} em estoque")
        else: print('OK')
    elif categoria == 'limpeza':
        if qtde < 30:
            print(f'Solicitar {produto} à equipe de compras, temos apenas {qtde} em estoque')
        else: print('OK')
    elif categoria == 'alimentos':
        if qtde < 50:
            print(f'Solicitar {produto} à equipe de compras, temos apenas {qtde} em estoque')
        else: print('OK')
else:
    print('Preencha todas as informações')
```
