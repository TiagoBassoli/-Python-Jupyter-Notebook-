#### datetime.timedelta()
A classe timedelta é usada para realizar operações com datas (adição e subtração).
#Ex: Tenho um projeto pra durar 387 dias, começa em 15 de fevereiro, o timedelta te ajuda a somar isso, leva em conta ano bissexto também

```python
from datetime import datetime, timedelta
data_inicio = datetime(2026, 2, 15)
duracao = 387
data_final = data_inicio + timedelta(days=duracao)
print(data_final)
```

```python
duracao_semanas = 15
data_final = data_inicio + timedelta(weeks=duracao_semanas)
print(data_final)
```

#### Oferecendo desconto para cliente com base na última compra

Analisando o tempo que passou desde a última transação do cliente, se for muito tempo o cliente pode ganhar um desconto. Vamos criar um script para ver quanto tempo passou desde a última compra do cliente (10 de maio de 2023). Se faz mais de 30 dias, ele receberá um desconto

```python
ultima_compra = datetime(2023,5,10)
data_atual = datetime.now() 

tempo_desdeacompra = (data_atual - ultima_compra)
print(tempo_desdeacompra)

if tempo_desdeacompra.days >= 30:
    print('Parabéns, você recebeu um desconto')
```

#### Data e hora em diferentes fusos horários
Uma empresa tem escritórios em São Paulo, Nova York e Tóquio. Crie um script que mostra a data e hora atuais nesses três fusos horários. Exiba também, se estes escritórios estão abertos ou fechados (9h às 17h) 

```python
from zoneinfo import ZoneInfo

data_hora_atual = datetime.now()

fuso_sp = ZoneInfo('America/Sao_Paulo')
fuso_ny = ZoneInfo('America/New_York')
fuso_tk = ZoneInfo('Asia/Tokyo')

data_horasp = data_hora_atual.astimezone(fuso_sp)
data_horany = data_hora_atual.astimezone(fuso_ny)
data_horatk = data_hora_atual.astimezone(fuso_tk)

print(f'Data/Hora em São Paulo: {data_horasp}')
print(f'Data/Hora em Nova York: {data_horany}')
print(f'Data/Hora em Tóquio: {data_horatk}')

#Agora pra ver se está aberto ou fechado
def verifica_hora(data_hora):
    if 9 <= data_hora.hour < 17:
        return 'aberto'
    else:
        return 'fechado'
    
print(f"Escritório de São Paulo está {verifica_hora(data_horasp)}")
print(f'Escritório de Nova York está {verifica_hora(data_horany)}')
print(f'Escritório de Tóquio está {verifica_hora(data_horatk)}')
```

#### Calculando a idade
Um usuário fornece sua data de nascimento no formato "dd/mm/aaaa". Crie um script Python que calcula a idade do usuário

```python
from datetime import datetime
data_nascimento_usuario = input("Informe sua data de nascimento (dd/mm/aaaa): ")
data_nascimento_usuario = datetime.strptime(data_nascimento_usuario, '%d/%m/%Y')
print(data_nascimento_usuario)

data_atual = datetime.now()
mes_atual = data_atual.month
dia_atual = data_atual.day

mes_nascimento = data_nascimento_usuario.month
dia_nascimento = data_nascimento_usuario.day

idade = data_atual.year - data_nascimento_usuario.year

if mes_nascimento > mes_atual:
    idade -= 1
elif mes_nascimento == mes_atual and dia_nascimento > dia_atual:
    idade -= 1
print(idade)
```
