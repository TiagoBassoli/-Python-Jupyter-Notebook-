#### Resumo do que fizemos

```python
import pandas as pd
#importando os arquivos
vendas_df = pd.read_csv(r'Contoso - Vendas - 2017.csv', sep=';', encoding='latin1')
produtos_df = pd.read_csv(r'Contoso - Cadastro Produtos.csv', sep=';', encoding='latin1')
lojas_df = pd.read_csv(r'Contoso - Lojas.csv', sep=';', encoding='latin1')
clientes_df = pd.read_csv(r'Contoso - Clientes.csv', sep=';',encoding='latin1')

# Ajustar nome das colunas
produtos_df.rename(columns={'ÿNome do Produto': 'Nome do Produto'}, inplace=True)
clientes_df.rename(columns={'ÿID Cliente': 'ID Cliente'}, inplace=True)
lojas_df.rename(columns={'ÿID Loja': 'ID Loja'}, inplace=True)

#limpando apenas as colunas que queremos
clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]

#mesclando e renomeando os dataframes
vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente').rename(columns={'E-mail': 'E-mail do Cliente'})
display(vendas_df)
```

### Qual a Loja que mais vendeu?

- Usaremos o .groupby para agrupar o nosso dataframe, de acordo com o que queremos (somando as quantidades de vendas, por exemplo)

```python
vendas_lojas = vendas_df.groupby("Nome da Loja").sum() #da pra colocar min, max, mean 
vendas_lojas = vendas_lojas[['Quantidade Vendida']] #Com os dois colchetes ele fica uma tabela bem feita, senão fica como se fossem dados fora de uma tabela
display(vendas_lojas)
```

- Agora precisamos pegar o maior valor. Temos 2 formas:
    1. Ordenar o dataframe em ordem decrescente de Quantidade Vendida
        - Método .sort_values
    2. Pegar o Maior valor diretamente
        - Métodos .max() e .idxmax()

```python
#ordenando o dataframe
vendas_lojas = vendas_lojas.sort_values('Quantidade Vendida', ascending=False) #sort_values ordena com base no valor
display(vendas_lojas)
#podemos plotar em um gráfico
vendas_lojas[:5].plot(figsize=(15,5),kind= 'bar')
```

```python
#pegando o maior valor e se indice
maior_valor = vendas_lojas['Quantidade Vendida'].max()
melhor_loja = vendas_lojas['Quantidade Vendida'].idxmax()
print(melhor_loja, maior_valor)
```

#### Filtrando Dataframes
Um dos grandes potenciais do Pandas é tratar condições

#### Primeiro, vamos aplicar uma função. Qual o % das vendas que foi devolvido?
- Para isso vamos somar as quantidades nas colunas correspondentes. Lembrando, o % vai ser: Total Devolvido / Total Vendido

```python
qtde_vendida = vendas_df['Quantidade Vendida'].sum()
qtde_devolvida = vendas_df['Quantidade Devolvida'].sum()
porcentagem_devolvido = qtde_devolvida / qtde_vendida
print(f'{porcentagem_devolvido:.2%} foi a porcentagem de vendas devolvidas')
```

#### Desafio: e se eu quisesse criar uma tabela apenas com as vendas da Loja Contoso Europe Online e que não tiveram nenhuma devolução. Quero criar essa tabela e saber quantas vendas são.
- Repare que nesse caso são duas condições, como fazemos isso?

```python
#tudo junto (1° maneira de fazer)
loja306_semdev = vendas_df[(vendas_df['ID Loja'] == 306) & (vendas_df['Quantidade Devolvida'] == 0)] #Tem de ter o parenteses para isolar as duas coisas, se não vai dar erro de ambiguidade (Vai tentar fazer as duas coisas ao mesmo tempo e travar)
display(loja306_semdev)
#separado (2° maneira de fazer)
loja306 = vendas_df['ID Loja'] == 306
devolucao_zero = vendas_df['Quantidade Devolvida'] == 0
df2_loja306_semdev = vendas_df[loja306 & devolucao_zero] #Fizemos cada condição separado e depois juntamos
display(df2_loja306_semdev)
```

### Integração Python + Excel
2 formas:
1. Pandas
- Mais usada no geral
- Trata o Excel como uma base de dados
- Faz o que quiser com o arquivo
- Pode desfazer a estrutura original do arquivo, caso queira editar (Por exemplo, se salvar com o mesmo nome do arquivo original, ele vai excluir os gráficos, porque ele os ignora)
2. Openpyxl
- Trata o Excel como uma planilha mesmo que possa ter várias coisas
- Edita 'como se fosse um VBA'
- Menos eficiente
- Mantém mais a estrutura original do arquivo, mas cuidado porque não necessariamente tudo, então tem que testar

#### Agora, imagina que a Loja Contoso Roma (ID 222), para tentar burlar o sistema de metas, diminuiu 1 da quantidade devolvida de todas as vendas que ela teve. Descobrindo isso, precisamos ajeitar a base

```python
from tqdm import tqdm
pbar = tqdm(total=len(vendas_df['ID Loja']), position=0, leave=True)
for i, id_loja in enumerate(vendas_df['ID Loja']):
    pbar.update()
    if id_loja == 222:
        vendas_df.loc[i,'Quantidade Devolvida'] += 1

display(vendas_df)
```

### Importação de módulos e Arquivos

```python
import pandas as pd
from openpyxl import Workbook, load_workbook

funcionarios_df = pd.read_csv(r"C:\Users\tiago\OneDrive\Área de Trabalho\Python\Arquivos CSV\CadastroFuncionarios.csv", sep=';', decimal= ',')
clientes_df = pd.read_csv(r"C:\Users\tiago\OneDrive\Área de Trabalho\Python\Arquivos CSV\CadastroClientes.csv", sep=';', decimal=',')
servicos_df = pd.read_excel(r"C:\Users\tiago\OneDrive\Área de Trabalho\Python\Bases de Dados\BaseServiçosPrestados.xlsx")

# vou excluir colunas que vão ser inúteis na Análise
funcionarios_df = funcionarios_df.drop(['Estado Civil', 'Cargo'], axis=1) # Axis 1 (Pra remover coluna), senão ia dar erro e ia remover as linhas
```

### 1 - Folha Salarial

```python
# vou ter que criar outra coluna pro Salario Total
funcionarios_df['Salario Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']
print('Total da Folha Salarial Mensal é de R${:,}'.format(funcionarios_df['Salario Total'].sum()))
```

### 2 - Faturamento da Empresa

```python
# vou ter que criar uma outra tabela só pro faturamento (já que as informações vem de duas tabelas diferentes)
faturamento_df = servicos_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente', 'Valor Contrato Mensal']], on= 'ID Cliente')
faturamento_df['Faturamento Total'] = faturamento_df['Tempo Total de Contrato (Meses)'] * faturamento_df['Valor Contrato Mensal']
print('O Faturamento Total da Empresa é de R${:,}'.format(sum(faturamento_df['Faturamento Total'])))
```

### 3 - % de Funcionários que fecharam contrato

```python
funcionarios_fecharam_contrato = len(servicos_df['ID Funcionário'].unique())
qtde_total_funcionarios = len(funcionarios_df['ID Funcionário'])
print('De todos os funcionários {:.2%} fecharam contrato'.format(funcionarios_fecharam_contrato / qtde_total_funcionarios))
```

### 4 - Qtde Contratos por Área

```python
# vamos ter que relacionar a coluna ID Funcionário da tabela de servicos e de funcionarios
contratos_area_df = servicos_df[['ID Funcionário']].merge(funcionarios_df[['ID Funcionário', 'Area']], on='ID Funcionário')
contratos_area_qtde = contratos_area_df['Area'].value_counts() # vai contar cada valor que estiver incluido nessa coluna de 'Area'
print(contratos_area_qtde)
contratos_area_qtde.plot(kind='bar')
```

### 5 - Funcionários por Área

```python
funcionarios_area = funcionarios_df['Area'].value_counts()
print(funcionarios_area)
funcionarios_area.plot(kind='bar')
```

### 6 - Ticket Médio

```python
ticket_medio = clientes_df['Valor Contrato Mensal'].mean()
print('Ticket Médio Mensal: R${:,.2f}'.format(ticket_medio))
```

Crie uma tabela com os anos 2016, 2017, 2018, 2019 e 2020 e o valor US$ de exportações

```python
tabela_anos_exportacao = tabela[['Year','US$ FOB']].groupby("Year").sum()
display(tabela_anos_exportacao)
tabela_anos_exportacao.plot(kind='bar')
```

#### Vamos formatar esse resultado para dólares
Só faça isso no final para exibição, você pode somar e fazer vários calculos enquanto ela não estiver formatada

```python
def formatar(valor):
    valor_formatado = f'US${valor:,}'
    return valor_formatado

tabela_anos_exportacao['US$ FOB'] = tabela_anos_exportacao['US$ FOB'].apply(formatar) #apply aplica a sua função para cada um dos valores da coluna
display(tabela_anos_exportacao)
```

Produtos com maior valor US$ de exportação

#### Em 2020, qual cidade mais exportou pra França?
Em valores US$

```python
tabela_2020 = tabela.loc[tabela['Year'] == 2020, :]

tabela_cidades_2020 = tabela_2020[['City', 'US$ FOB']].groupby('City').sum()
tabela_cidades_2020 = tabela_cidades_2020.sort_values(by= 'US$ FOB', ascending=False) #Ordenar valores de 'US$ FOB' decrescente

tabela_cidades_2020['US$ FOB'] = tabela_cidades_2020['US$ FOB'].apply(formatar)
display(tabela_cidades_2020)
```

#### Quais os produtos mais exportados (em US$) que as 2 maiores cidades (em exportação em 2020) exportaram?

```python
cidade = tabela_cidades_2020.index[0]

tabela_cidade = tabela_2020.loc[tabela_2020['City'] == cidade, :]

tabela_produtos_cidade = tabela_cidade[['SH2 Description', 'US$ FOB']].groupby('SH2 Description').sum()

tabela_produtos_cidade = tabela_produtos_cidade.sort_values(by= 'US$ FOB', ascending=False)
tabela_produtos_cidade['US$ FOB'] = tabela_produtos_cidade['US$ FOB'].apply(formatar)

print(f"Esses foram os produtos mais exportados da cidade de {cidade}")
display(tabela_produtos_cidade)
```
