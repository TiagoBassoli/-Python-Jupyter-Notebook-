#### Seaborn é mais adequado trabalhando com chave e valor (Pode ser Dataframe ou dicionário)

```python
import seaborn as sns

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

dados = {'Meses': meses, 'Vendas': vendas}

sns.lineplot(x='Meses',y='Vendas', data=dados) # passei dados com chave e valor, eles foram adicionados como n° e título do gráfico
```

#### Estou lidando com um sistema de eixos do Matplotlib 

```python
linha = sns.lineplot(x='Meses',y='Vendas', data=dados)
print(type(linha))
```

#### Isso quer dizer, que tudo que aprendemos anteriormente continua válido (Sobre trabalhar com eixos, 'axs')

```python
linha = sns.lineplot(x='Meses',y='Vendas', data=dados)
linha.set_title('Vendas ao longo do ano') #repare, funciona como o 'axs.set_title()'
```

#### É possível também passar um sistema de eixos do próprio Matplotlib pro Seaborn

```python
import matplotlib.pyplot as plt

fig, eixo = plt.subplots()

sns.lineplot(x='Meses',y='Vendas', data=dados, ax=eixo)

eixo.set_title('Vendas ao longo do ano')
plt.show()
```

#### Criando gráfico de barras no Seaborn

```python
import seaborn as sns

cidades = ['Brasília', 'Rio de Janeiro', 'São Paulo', 'Curitiba', 'Manaus']
temperaturas = [27, 29, 25, 23, 31]

tituloeixos = {'Cidades': cidades, 'Temperaturas (C°)': temperaturas}

barras = sns.barplot(x='Cidades', y='Temperaturas (C°)', data=tituloeixos); # ';' evita que apareça um código antes do gráfico
barras.set_title('Temperaturas médias anuais'); # ';' evita que apareça um código antes do gráfico
```

#### Temas do Seaborn
No Seaborn, um tema é a combinação de estilo, contexto e paleta de cores que juntos definem a aparência de um gráfico. Mais bem trabalhado que o Matplotlib e feito para melhor apresentação de dados

```python
import seaborn as sns

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

dados = {'Meses': meses, 'Vendas': vendas}

sns.lineplot(x='Meses',y='Vendas',data= dados);
```

#### Diferentes tipos de estilos, contextos e paletas encontrados no Seaborn

```python
estilos = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
contextos = ['paper', 'notebook', 'talk', 'poster']
paletas = ['deep', 'muted', 'bright', 'pastel', 'dark', 'colorblind'] #além das do matplotlib
```

#### O tema padrão usa o estilo 'darkgrid'

```python
sns.set_theme() 

sns.lineplot(x='Meses',y='Vendas',data= dados);
```

#### Todos os estilos abaixo:

```python
sns.set_theme(style='darkgrid')

sns.lineplot(x='Meses',y='Vendas',data= dados);
```

```python
sns.set_theme(style='whitegrid')

sns.lineplot(x='Meses',y='Vendas',data= dados);
```

```python
sns.set_theme(style='white')

sns.lineplot(x='Meses',y='Vendas',data= dados);
```

```python
sns.set_theme(style='dark')

sns.lineplot(x='Meses',y='Vendas',data= dados);
```

```python
sns.set_theme(style='ticks')

sns.lineplot(x='Meses',y='Vendas',data= dados);
```

#### O contexto padrão do Seaborn é o notebook para trabalhar com o Jupyter Notebook

```python
sns.set_theme(style='darkgrid', context= 'notebook')

sns.lineplot(x='Meses',y='Vendas',data= dados);
```

#### Todos os contextos abaixo:

```python
sns.set_theme(style='darkgrid', context= 'paper') #paper fica no estilo de um artigo

sns.lineplot(x='Meses',y='Vendas',data= dados);
```

```python
sns.set_theme(style='darkgrid', context= 'talk') #talk é feito para palestras

sns.lineplot(x='Meses',y='Vendas',data= dados);
```

```python
plt.figure(figsize=(10,7))
sns.set_theme(style='darkgrid', context= 'poster')
sns.lineplot(x='Meses',y='Vendas',data= dados);
```

Como vimos, o 'context' elabora o gráfico para diferentes contextos a serem trabalhados, como um artigo, uma palestra, poster e notebook

```python
sns.set_theme(style='darkgrid', context= 'notebook', palette='Pastel2')
sns.lineplot(x='Meses',y='Vendas',data= dados);
```

#### Análise de Dados com Pandas e Seaborn

Se o valor de id_cliente na coluna 'unique' desse diferente do 'count' teriamos um problema. Grande parte das colunas de unique tem 2 valores porque são colunas de 'sim' ou 'não'. #### Vamos tirar colunas que não vão fazer a diferença em nossa análise

```python
df_churn = df_churn.drop(columns='id_cliente')
df_churn.head()
```

#### Separar as colunas em numericas e categoricas

```python
colunas_numericas = df_churn.select_dtypes(include='number').columns



colunas_categoricas = df_churn.select_dtypes(exclude='number').columns

colunas_categoricas = colunas_categoricas.drop('churn') # só vou tirar o churn, porque mesmo não sendo número, queremos entende-la, sabendo como as outras colunas influenciam ela

df_churn.head()
```

#### Histograma

```python
df_churn[colunas_numericas].hist(layout=(1,3), figsize=(14,6));
```

```python
sns.set_theme(style= 'dark', palette='bright', context='notebook')

sns.histplot(x='meses_na_empresa', data=df_churn, hue= 'contrato');
```

#### Vamos ver as barras sobrepostas agora e não encima uma da outra

```python
sns.histplot(x='meses_na_empresa', data=df_churn, hue= 'contrato', multiple='stack');
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(14,6))

for i, coluna in enumerate(colunas_numericas):
    sns.histplot(x=coluna, data=df_churn, hue='contrato', multiple='stack', ax = axs[i])

plt.show
```
