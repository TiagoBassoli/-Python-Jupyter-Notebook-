#### Programa que consiga categorizar a idade das equipes de uma empresa. Faça um programa que peça para n pessoas a sua idade, ao final o programa devera verificar se a média de idade da equipe varia entre 0 e 25 (jovem) ,26 e 60 (sênior) e maior que 60 (idosa); e então, dizer se a equipe é jovem, sênior ou idosa, conforme a média calculada.

```python
qtde = int(input(f'Informe quantas pessoas serão: '))

total = 0

for i in range(qtde):
    idade = float(input(f'Informe a {i+1}ª idade: '))
    total += idade

media = total/qtde

if 0 <= media <= 25:
    print('A equipe é jovem')
elif 25 < media <= 60:
    print('A equipe é adulta')
elif 60 < media:
    print('A equipe é idosa')
```
