#### Iterable

É uma estrutura que armazena dados que pode ser "iterada" ou seja, que é possivel fazer um loop como for dentro dela. É como se fosse um tipo de lista de coisas que você pode ir olhando cada elemento

#### Principais Iterables
- Listas
- Strings
- Tuplas
- Dicionários

```python
#Podemos usar o Range de três formas nos Iterables, abaixo o uso mais comum no for:
produtos = ['arroz', 'feijão', 'macarrão', 'atum', 'azeite']
estoque = [50, 100, 20, 5, 80]

for i in range(5):
    print('{}: {} unidades'.format(produtos[i], estoque[i]))
```

```python
#Aqui um uso de começo e fim, ele não printa o último número
for i in range(1,10):
    print(i)
```

- Modelo de gestão Jack Weich da G&E

1. Classe A: 10% melhor
2. Classe B: 80% mantém/busca melhorar
3. Classe C: 10% pior

```python
#Quem são os funcionários Classe B?

funcionarios = ['Maria','José','Antônio','João','Francisco','Ana','Luiz', 'Paulo', 'Carlos', 'Manoel', 'Pedro']
vendas = [2750, 1900, 1500, 1200, 1111, 1100, 999, 900, 880, 870, 800]

print('Funcionários Classe B')
for i in range(1, 10):
    print('{}: fez {} vendas'.format(funcionarios[i], vendas[i]))
```

```python
#range com passo
for i in range(0, 100, 10):
    print(i)
```

- Aplicação bem útil
Quantos clientes tivemos na loja?

```python
cpf_clientes = ['902.820.610-88', '578.434.480-39', '578.434.480-39', '276.211.530-24', '288.537.890-50', '288.537.890-50', '959.221.250-36', '194.270.390-27', '596.497.640-97' ]
numero_clientes = len(cpf_clientes) 
print(numero_clientes) #Temos nossa lista, armazenamos o comprimento dela em uma variável e printamos "(9)"

clientes_semduplicatas = set(cpf_clientes)
print(clientes_semduplicatas) #Criamos um set pra eliminar os valores duplicados e armazenamos em uma variável

clientes_unicos = list(clientes_semduplicatas)
print('Temos {} clientes em nossa loja'.format(len(clientes_unicos))) #Depois convertemos pra uma lista, só que agora sem duplicatas, e printamos nossos clientes
```
